# 유사한 단어 찾기 게임

1. 사전 학습된 모델 또는 적절한 데이터셋을 찾는다
2. 워드 임베딩 모델을 학습시킨다
3. 단어 유사도가 0.8 이상인 A, B를 랜덤 추출한다
4. A, B와 대응되는 C를 추출한다
5. D를 입력받는다

=><br>
A:B = C:D 관계에 대응하는 D를 찾는 게임을 만든다.<br>
ex) A: 산, B: 바다, C: 나무, D: 물

**<출력 예시>**

관계 [ 수긍 : 추락 = 대사관 : ? ]<br>
모델이 예측한 가장 적합한 단어: 잠입<br>
당신의 답변과 모델 예측의 유사도: 0.34<br>

```python
import os
import json
import glob
import pandas as pd

filename_list = []
opinions_list = []
sentence_list = []

path = 'data/Training/01.원천데이터'
folders = os.listdir(path)
for folder in folders:
    folder_path = os.path.join(path,folder)
    all_files = glob.glob(os.path.join(folder_path, '*training.json'))
    print(all_files)
    filename_list.extend(all_files)

for file in filename_list:

    with open(file, 'r') as f:
        json_data = json.load(f)
    sentence_list.extend(json_data)

df = pd.DataFrame(sentence_list)
display(df)
```

```python
dataset = df[['id', 'title', 'field', 'sentence']]
dataset
```

```python
import re
import string
import pandas as pd
from konlpy.tag import Okt

# 결측치 전처리
dataset.isnull().sum()
dataset = dataset.dropna(how='any')
```

```python
def load_stopwords(filepath):
    with open(filepath, 'r', encoding='UTF-8') as f:
        stopwords = [line.strip() for line in f]
    return stopwords

def preprocess(sentence, okt):

    # 영문, 특수문자, 이모지, 공백 제거
    sentence = re.sub(r'[^0-9가-힣ㄱ-ㅎㅏ-ㅣ\s]', '', sentence)

    # 어근 분리 및 토큰화
    tokens = okt.morphs(sentence, stem =True)

    # 불용어 제거
    ko_stopwords = load_stopwords('data/ko_stopwords.txt')

    cleaned_tokens = [token for token in tokens if token not in ko_stopwords]

    return cleaned_tokens
```

```python
from konlpy.tag import Okt
from tqdm import tqdm       # 진행도 시각화

okt = Okt()
preprocessed_data = []

for sentence in tqdm(dataset['sentence']):
    tokens = preprocess(sentence, okt)
    preprocessed_data.append(tokens)
```

```python
from gensim.models import Word2Vec

model = Word2Vec(
    sentences=preprocessed_data,
    vector_size=200,
    window=10,
    min_count=5,
    sg=1
)

model.wv.vectors.shape
```

```python
model.wv.save_word2vec_format('ckpt/pretrained_200.bin', binary=False)
```

```python
from gensim.models import KeyedVectors

loaded_wv = KeyedVectors.load_word2vec_format('ckpt/pretrained_200.bin', binary=False)
loaded_wv.vectors.shape
```

```python
import random
import numpy as np

words = list(loaded_wv.index_to_key)

def find_similar_pair(min_similarity=0.8, max_trials=1000):
    for _ in range(max_trials):
        A = random.choice(words)
        similar_words = [
            word for word, sim in loaded_wv.most_similar(A, topn=50)
            if sim >= min_similarity and word != A
        ]
        if len(similar_words) >= 2:  # 최소 2개 이상 있어야 B, C 선택 가능
            B, C = random.sample(similar_words, 2)
            return A, B, C
    return None, None, None

A, B, C = find_similar_pair()
print("A:", A, "B:", B, "C:", C)

# Word2Vec 벡터 연산
D_vector = loaded_wv[B] - loaded_wv[A] + loaded_wv[C]
D_candidates = loaded_wv.similar_by_vector(D_vector, topn=5)
D_candidates_words = [(word, sim) for word, sim in D_candidates if (word != A and word != B and word != C)]
predicted = max(D_candidates_words, key=lambda x: x[1])[0]

print(f'모델이 예측한 가장 적합한 단어: {predicted}')
```

```python
print(f'관계 [ {A} : {B} = {C} : ? ]')
D = input('당신의 답변을 입력하세요 : ')

# 모델에 있는 단어인지 확인
if D in loaded_wv:
    sim = loaded_wv.similarity(D, predicted)
    print(f'모델이 예측한 가장 적합한 단어: {predicted}')
    print(f'당신이 입력한 단어: {D}')
    print(f'당신의 답변과 모델 예측의 유사도: {sim:.3f}')
else:
    print(f"입력한 단어 '{D}'는 모델 사전에 없습니다.")
    print(f'모델이 예측한 가장 적합한 단어: {predicted}')
    print('유사도를 계산할 수 없습니다. 다른 단어를 시도해보세요.')
```
